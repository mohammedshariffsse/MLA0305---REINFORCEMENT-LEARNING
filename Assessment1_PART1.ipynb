## Markov Decision Process (MDP) Grid World Simulation

## Code 1: MDP Grid World Simulation


```python
grid_size = 5

start = (0, 0)
goal = (4, 4)

state = start
step = 1

print("Starting State:", start)
print("Goal State:", goal)

while state != goal:

    row, col = state

    possible_moves = []

    if row < grid_size - 1:
        possible_moves.append((row + 1, col))

    if col < grid_size - 1:
        possible_moves.append((row, col + 1))

    next_state = min(
        possible_moves,
        key=lambda x: abs(goal[0]-x[0]) + abs(goal[1]-x[1])
    )

    reward = -2

    if next_state == goal:
        reward = 50

    print("\nStep:", step)
    print("Current State:", state)
    print("Next State:", next_state)
    print("Reward:", reward)

    state = next_state
    step += 1
```

## Code 2: Agent Path Visualization

```python
import matplotlib.pyplot as plt

x = [0,0,1,2,3,4,4,4,4]
y = [0,1,1,1,1,1,2,3,4]

plt.figure(figsize=(6,6))

plt.plot(y, x, marker='o')

plt.scatter(0,0,s=200,label="Start")
plt.scatter(4,4,s=200,label="Goal")

plt.grid(True)

plt.title("Agent Path in Grid World")
plt.xlabel("Column")
plt.ylabel("Row")

plt.legend()

plt.show()
```

## Code 3: MDP Workflow Diagram

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

nodes = [
    "State",
    "Action",
    "Environment",
    "Reward",
    "Policy",
    "Optimal Action"
]

G.add_nodes_from(nodes)

G.add_edges_from([
    ("State","Action"),
    ("Action","Environment"),
    ("Environment","Reward"),
    ("Reward","Policy"),
    ("Policy","Optimal Action")
])

plt.figure(figsize=(8,5))

pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=3000
)

plt.title("MDP Workflow Diagram")

plt.show()
```

## Interpretation

The agent moves through the grid world by selecting actions that bring it closer to the goal state. The workflow diagram illustrates how states, actions, rewards, and policies interact in a Markov Decision Process.
